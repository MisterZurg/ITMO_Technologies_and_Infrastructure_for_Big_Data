```python
npy_file = np.load("../../rdd_presentation_test_backup.npy")
# print(npy_file.shape)
df = pd.DataFrame(npy_file)
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
from PIL import Image
import imagehash

def get_hash(data):
    im = Image.fromarray((np.array(data).reshape((228, 228, 3))*255).astype(np.uint8))
    hash = imagehash.average_hash(im)
    return hash

df["hash"] = df.apply(get_hash, axis=1)
df.head()
```

```python
print(len(df) - len(pd.unique(df["hash"])), "duplicates")
df["dupe"] = df["hash"].duplicated()
df_dupes = df[df["dupe"]].drop(["hash", "dupe"], axis=1)
df_dupes.head()
```

```python
for i in range(len(df_dupes)):
    cur_im = np.array(df_dupes.iloc[i]).reshape((228, 228, 3))
    plt.subplot(1, len(df_dupes), i+1)
    plt.imshow(cur_im, interpolation='nearest')
    plt.axis("off")
```

```python
from PIL import ImageStat
import math

def brightness(flat_arr):
   im = Image.fromarray((np.array(flat_arr).reshape((228, 228, 3))*255).astype(np.uint8))
   stat = ImageStat.Stat(im)
   r,g,b = stat.mean
   return math.sqrt(0.241*(r**2) + 0.691*(g**2) + 0.068*(b**2))
```

```python
df_uniq = df[df["dupe"] == False]
df_uniq = df_uniq.drop(["dupe", "hash"], axis=1)

metrics = df_uniq.copy()
metrics["br"] = df_uniq.apply(brightness, axis=1)
```

```python
brs = np.array(metrics["br"]);
plt.rcParams["figure.figsize"] = (10, 5)
plt.hist(brs, bins="auto", color="black", alpha=0.4);
plt.grid();
plt.title("Brightness distribution")
plt.xlabel("Brightness");
plt.ylabel("Frequency");
```

```python
def rr(flat_arr):
   im = Image.fromarray((np.array(flat_arr).reshape((228, 228, 3))*255).astype(np.uint8))
   stat = ImageStat.Stat(im)
   r,_,_ = stat.mean
   return r

def gg(flat_arr):
   im = Image.fromarray((np.array(flat_arr).reshape((228, 228, 3))*255).astype(np.uint8))
   stat = ImageStat.Stat(im)
   _,g,_ = stat.mean
   return g

def bb(flat_arr):
   im = Image.fromarray((np.array(flat_arr).reshape((228, 228, 3))*255).astype(np.uint8))
   stat = ImageStat.Stat(im)
   _,_,b = stat.mean
   return b

n = len(df_uniq.iloc[0])
metrics["red"] = df_uniq.apply(rr, axis=1)
metrics["green"] = df_uniq.apply(gg, axis=1)
metrics["blue"] = df_uniq.apply(bb, axis=1)

metrics = metrics[["br", "red", "green", "blue"]]
metrics["is_bw"] = metrics.apply(lambda x: x["red"] == x["green"] and x["red"] == x["blue"], axis=1)
metrics.head()
```

```python
for col in ["red", "green", "blue"]:
    brs = np.array(metrics[col]);
    plt.hist(brs, bins="auto", alpha=0.25, color=col);

plt.grid();
plt.title("Color distribution")
plt.xlabel("RGB channel intensity");
plt.ylabel("Frequency");
```
